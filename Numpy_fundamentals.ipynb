Stacking

```python
## Horizontal stacking

a4 = np.arange(12).reshape(3, 4)
a5 = np.arange(12, 24).reshape(3, 4)

print(a4)
print()
print(a5)

```

```python
np.hstack((a4, a5))
```

```python
np.hstack((a4, a5, a4))
```

```python
## Vertical Stacking

np.vstack((a4, a5))
```
